```python
# Install library yang diperlukan
!pip install ultralytics roboflow

# Import library
from roboflow import Roboflow
from ultralytics import YOLO
import cv2
from google.colab.patches import cv2_imshow
from google.colab import files
import os

# Unduh dataset dari Roboflow
rf = Roboflow(api_key="EBbdoZllvDs3qEGG3t5F")
project = rf.workspace("elis-xh6v9").project("deteksi-gambar-setrika")
version = project.version(1)
dataset = version.download("yolov8")

# Path ke file data.yaml
data_yaml_path = f"{dataset.location}/data.yaml"

# Latih model YOLOv8 dengan augmentasi
model = YOLO("yolov8n.pt")  # Gunakan YOLOv8n sebagai model awal
model.train(data=data_yaml_path, epochs=20, imgsz=640, augment=True)  # Aktifkan augmentasi

# Periksa keberadaan file best.pt setelah pelatihan
trained_model_path = 'runs/detect/train/weights/best.pt' if os.path.exists('runs/detect/train/weights/best.pt') else 'yolov8n.pt'

# Gunakan model yang sudah dilatih untuk deteksi
model = YOLO(trained_model_path)

# Fungsi untuk deteksi gambar
def detect_image(image_path):
    # Jalankan deteksi pada gambar
    results = model(image_path, conf=0.25)  # Atur ambang batas kepercayaan
    detections = results[0]  # Ambil hasil deteksi pertama

    # Ambil gambar asli
    img = cv2.imread(image_path)

    # Jika tidak ada objek yang terdeteksi
    if len(detections.boxes) == 0:
        print("Tidak ada objek yang terdeteksi.")
        cv2_imshow(img)
        return

    # Iterasi hasil deteksi
    for box in detections.boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())  # Koordinat bounding box
        confidence = float(box.conf[0])  # Skor kepercayaan
        label = detections.names[int(box.cls[0])]  # Label deteksi

        # Menggambar kotak pada gambar (warna hijau)
        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)

        # Menambahkan teks label di atas kotak
        cv2.putText(img, f"{label} {confidence:.2f}",
                    (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

    # Menampilkan gambar yang sudah diberi kotak dan label
    cv2_imshow(img)

    # Simpan gambar yang sudah diberi label
    detected_image_path = f"detected_{os.path.basename(image_path)}"
    cv2.imwrite(detected_image_path, img)
    print(f"Gambar dengan deteksi disimpan sebagai: {detected_image_path}")

# Unggah gambar langsung ke Colab
print("Unggah gambar untuk deteksi:")
uploaded = files.upload()  # Pilih gambar yang ingin diunggah

# Setelah gambar diunggah, jalankan deteksi
for filename in uploaded.keys():
    image_path = f"/content/{filename}"  # Path ke gambar yang diunggah
    print(f"Deteksi gambar: {filename}")
    detect_image(image_path)
```
